```python
# Plot the measles cases in Afghanistan by year using seaborn
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid")
sns.lineplot(data=df[df['country'] == 'Afghanistan'], x='year', y='measles_total', marker='o')
plt.title('Measles Cases in Afghanistan by Year')
plt.xlabel('Year')
plt.ylabel('Number of Cases')
```

```python
mdf = pd.read_csv('cases_month.csv')
mdf.head()
```

```python
# populate some previous season stats to compare
g = df.groupby(["iso3", "year"])["measles_total"].sum()
joined = (
    g.to_frame()
    .assign(current_measles=g)
    .join(g.groupby(level=[0]).shift().to_frame(), lsuffix="", rsuffix="_")
    .rename(columns={"measles_total_": "previous_measles"})
)
joined.update(
    joined[["current_measles", "previous_measles"]].mask(
        joined["previous_measles"].isna()
        | joined["previous_measles"].eq(joined["current_measles"]),
        0,
    )
)
agg_df = df.groupby(["iso3", "year"]).agg(
    {
        "measles_total": "sum",
        "country": "first",
        "total_population": "max"
    }
)
agg_df["previous_measles"] = joined["previous_measles"].astype("Int64")
agg_df['absolute_change'] = agg_df['measles_total'] - agg_df['previous_measles']
agg_df['change_per_population'] = abs(agg_df['absolute_change'] / agg_df['total_population'])
agg_df.head()
agg_df.sort_values(by='change_per_population', ascending=False).head(20)
```

```python
import matplotlib.dates as mdates
import numpy as np
mongolia_outbreak_df = mdf[(mdf['iso3'] == 'MNG') & (mdf['year'].isin([2014, 2015, 2016, 2017]))]
# reset indext to month and date
mongolia_outbreak_df['date'] = pd.to_datetime(mongolia_outbreak_df['year'].astype(str) + '-' + mongolia_outbreak_df['month'].astype(str) + '-01')
mongolia_outbreak_df.set_index('date', inplace=True)
# plot the measles cases in Mongolia by month using seaborn
ax = sns.lineplot(data=mongolia_outbreak_df, x=mongolia_outbreak_df.index, y='measles_total', marker='o')
# Add ticks only for each year
ax.xaxis.set_major_locator(mdates.YearLocator())
years_fmt = mdates.DateFormatter('%Y')
ax.xaxis.set_major_formatter(years_fmt)
print(mongolia_outbreak_df.iloc[0, 3])
datemin = np.datetime64(mongolia_outbreak_df.index.array[0], 'Y')
datemax = np.datetime64(mongolia_outbreak_df.index.array[-1], 'Y') + np.timedelta64(1, 'Y')

ax.set_xlim(datemin, datemax)
# plt.xlabel('Year')
plt.ylabel('Measles Cases')
plt.title('Measles Cases in Mongolia by Month (2014-2017)')
plt.legend(title='Year')
```
